# Data Cleaning


file download and have an initial look at the dataframe.

```python
import requests
import pandas as pd
import os
from google.colab import files

def download_and_process_excel(url):
    filename = os.path.basename(url)
    response = requests.get(url)

    with open(filename, "wb") as file:
        file.write(response.content)


    df = pd.read_excel(
        filename,
        sheet_name="Incident",
    )

    # Drop any completely empty rows or columns
    df.dropna(how="all", axis=0, inplace=True)
    df.dropna(how="all", axis=1, inplace=True)

    return df
df = download_and_process_excel("https://github.com/1zpzz/QM2---school-shooting-/raw/refs/heads/main/Public%20v3.1%20K-12%20School%20Shooting%20Database%20(11%2025%202024).xlsx")
df
```

```python
print(df.columns)
```

filter out some columns

```python
columns_to_keep = ['Year', 'Date', 'School', 'City', 'State', 'Victims_Killed', 'Victims_Wounded', 'Number_Victims', 'Shooter_Killed','Summary','LAT', 'LNG']
df1 = df[columns_to_keep].copy()
df1['Year'] = df1['Year'].astype(str)
df1
```

map the state and their abbreviation and adjust the data formate for the "Date"

```python
df1.loc[:, 'Date'] = pd.to_datetime(df1['Date'], errors='coerce').dt.strftime('%Y-%m-%d')
df1.loc[:, 'State_Name'] = df1['State'].map(us_state_abbrev)
```

# Visualisation

interactive map

```python
import plotly.express as px
ordered_years = [str(y) for y in range(1966, 2025)]
fig = px.scatter_mapbox(
    df1,
    lat='LAT',
    lon='LNG',
    hover_name='School',
    hover_data={'Year':False,
                'Date':True,
                'City':True,
                'State':False,
                'State_Name':True,
                'Victims_Killed':True,
                'Victims_Wounded':True,
                'Number_Victims':True,
                'Shooter_Killed':True,
                'Summary':True,
                'LAT':False,
                'LNG':False
                },
    color='Victims_Killed',
    size='Number_Victims',
    size_max=30,
    zoom=3.7,
    center={"lat": 37.0902, "lon": -95.7129},
    #mapbox_style="carto-darkmatter",
    mapbox_style="open-street-map",
    animation_frame='Year',
    category_orders={"Year": ordered_years},
)

fig.update_layout(
    title='K-12 School Shooting Incidents by Year',
    margin=dict(l=0, r=0, t=0, b=0),
    paper_bgcolor="#aad3df",
    plot_bgcolor="#aad3df",
)

fig.show()
```

run the cell below to download the map as html file.

```python
fig.write_html("school_shooting_map.html")
files.download('school_shooting_map.html')
```

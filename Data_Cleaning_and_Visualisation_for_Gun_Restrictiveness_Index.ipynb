# Data Cleaning

Download the data from github and drop the empty cells.

```python
!pip install us
import requests
import pandas as pd
import os
from google.colab import files
import plotly.express as px
import us

def download_and_process_excel(url):
    filename = os.path.basename(url)
    response = requests.get(url)

    with open(filename, "wb") as file:
        file.write(response.content)


    df = pd.read_excel(
        filename,
        sheet_name="State Average gun law restricti",
    )

    # Drop any completely empty rows or columns
    df.dropna(how="all", axis=0, inplace=True)
    df.dropna(how="all", axis=1, inplace=True)

    return df
df = download_and_process_excel("https://github.com/1zpzz/QM2---school-shooting-/raw/refs/heads/main/Gun_Law.xlsx")
df
```

drop the last row since it does not contain any index.

```python
df = df[:-1]
df.loc[:, "Abbreviation"] = df["State"].map(lambda name: us.states.lookup(name).abbr)
df
```

melt the dataframe so it fits the requirement for the plotly.

```python
df_long = df.melt(
    id_vars=["State", "Abbreviation"],
    var_name="Year",
    value_name="Score"
)
df_long['Score'] = pd.to_numeric(df_long['Score'], errors='coerce')
df_long
```

# Visualisation

Interactive map

```python
fig = px.choropleth(
    df_long,
    locations="Abbreviation",
    locationmode="USA-states",
    color="Score",
    animation_frame="Year",
    #color_continuous_scale=[(0.0, "red"), (1.0, "white")],
    color_continuous_scale=px.colors.sequential.Reds_r,
    scope="usa",
    hover_data={
        "State": True,
        "Score": True,
        "Year": False,
        "Abbreviation": False,
    }
)
fig.update_geos(
    showlakes=False,
    showrivers=False,
    showocean=False
)
fig.add_annotation(
    text="The Restrictiveness Index decrease for each <br> permissive or absent law in a category <br> and increase for each restrictive law.",
    showarrow=False,
    x=1.02,
    xref='paper',
    xanchor='right',
    y=0.8,
    yref='paper',
    yanchor='middle',
    align='right',
    font=dict(size=8)
)
fig.update_layout(
    title="Restrictiveness Index by State from 2006 - 2022",
    margin={"r":0,"t":40,"l":0,"b":0}
)

fig.show()
```

```python
fig.write_html("gunlaw_map.html")
files.download('gunlaw_map.html')
```
